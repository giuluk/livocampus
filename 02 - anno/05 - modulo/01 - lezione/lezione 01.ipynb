```python
# Importing necessary libraries
import sympy as sp

# Importing matplotlib.pyplot is generally not required for SymPy plots.
# However, in some online notebook environments, it ensures that plots are rendered correctly.
import matplotlib.pyplot as plt
```

```python
# Defining the variables
x = sp.Symbol('x')
y = sp.Symbol('y')
```

```python
# Defining the expressions for the equations
expr1 = 3*x + 2*y -1
expr2 = x**2 + x*y + 3*y**2 - 6

# Defining the equations
eqzn1 = sp.Eq(expr1, 0)
eqzn2 = sp.Eq(expr2, 0)
```

```python
# Plotting the implicit equations
grfc1 = sp.plot_implicit(
    eqzn1, 
    aspect_ratio=(1,1), 
    line_color='darkblue', 
    show=False
)
grfc2 = sp.plot_implicit(
    eqzn2, 
    aspect_ratio=(1,1), 
    line_color='red', 
    show=False
)
```

```python
# Displaying the individual plots
grfc1.show() # Show the plot for the linear equation
grfc2.show() # Show the plot for the quadratic equation
```

```python
# Combining the plots into a grid
grfc = sp.plotting.PlotGrid(1, 2, grfc1, grfc2, show=False)

# Displaying the combined plot
grfc.show()
```
