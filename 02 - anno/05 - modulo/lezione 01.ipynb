```python
# Importing necessary libraries
import sympy as sp
```

```python
# Defining the variables
x = sp.Symbol('x')
y = sp.Symbol('y')
```

```python
# Defining the expressions for the equations
expr1 = 3*x + 2*y -1
expr2 = x**2 + x*y + 3*y**2 - 6

# Defining the equations
eqzn1 = sp.Eq(expr1, 0)
eqzn2 = sp.Eq(expr2, 0)
```

```python
# Plotting the implicit equations
grfc1 = sp.plot_implicit(
    eqzn1, 
    aspect_ratio=(1,1), 
    line_color='darkblue', 
    show=False
)
grfc2 = sp.plot_implicit(
    eqzn2, 
    aspect_ratio=(1,1), 
    line_color='red', 
    show=False
)
```

```python
grfc1.extend(grfc2) # Merging the two plots
grfc1.show() # Show the merged plot
```
