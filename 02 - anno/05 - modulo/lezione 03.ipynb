```python
# Importing necessary libraries
import sympy as sp
```

```python
# Defining the variables
x = sp.Symbol('x')
y = sp.Symbol('y')
```

```python
# Defining the expressions for the equations
expr_lin1 = 3*x -y -4
expr_lin2 = 2*x +2*y +1
expr_lin3 = 2*x -y -6
expr_quad = 2*x**2 +3*x*y +3*y**2 -5

# Defining the equations
eqzn_lin1 = sp.Eq(expr_lin1, 0)
eqzn_lin2 = sp.Eq(expr_lin2, 0)
eqzn_lin3 = sp.Eq(expr_lin3, 0)
eqzn_quad = sp.Eq(expr_quad, 0)

# Defining system equations
syst_lin  = (eqzn_lin1, eqzn_lin2)
syst_quad = (eqzn_lin3, eqzn_quad)
```

```python
# Solving the system of equations
sol1 = sp.linsolve(syst_lin, (x, y))
sol2 = sp.nonlinsolve(syst_quad, (x, y))
```

```python
# Displaying the solution for the linear system
sol1
```

```python
# Displaying the solution for the nonlinear system
sol2
```

```python
# Plotting the implicit equations
grfc_lin1 = sp.plot_implicit(
    eqzn_lin1, 
    aspect_ratio=(1,1), 
    title='Posizione reciproca tra rette',
    line_color='darkblue', 
    show=False
)
grfc_lin2 = sp.plot_implicit(
    eqzn_lin2, 
    aspect_ratio=(1,1), 
    line_color='red', 
    show=False
)
grfc_lin3 = sp.plot_implicit(
    eqzn_lin3, 
    aspect_ratio=(1,1), 
    title='Posizione reciproca tra retta e conica',
    line_color='purple', 
    show=False
)
grfc_quad = sp.plot_implicit(
    eqzn_quad, 
    aspect_ratio=(1,1), 
    line_color='darkgreen', 
    show=False
)
```

```python
# Mergin and show the two plots
grfc_lin1.extend(grfc_lin2) 
grfc_lin1.show()

grfc_lin3.extend(grfc_quad) 
grfc_lin3.show()
```
